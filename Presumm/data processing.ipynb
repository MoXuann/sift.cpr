```python
import spacy
from spacy.lang.en import English
import json
import en_core_web_md
```

```python
f = open("iphone.txt", "r")
```

```python
f1 = f.readlines()
```

```python
cat_sentence = ""
for x in f1:
    if len(x) < 2:
        continue
    x = x.strip()
    cat_sentence += x
```

```python
nlp = English()  # just the language with no model
sentencizer = nlp.create_pipe("sentencizer")
nlp.add_pipe(sentencizer)
splited = []
doc = nlp(cat_sentence)
for sent in doc.sents:
    splited.append(str(sent))
```

```python
nlp = en_core_web_md.load()
src_list = []
for x in splited:
    tokens = [word.text for word in nlp(x)]
    src_list.append(tokens)
```
