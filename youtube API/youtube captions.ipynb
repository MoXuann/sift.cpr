```python
from youtube_transcript_api import YouTubeTranscriptApi
```

```python
video_id = "oL58Ognb_BY"
```

```python
result = YouTubeTranscriptApi.get_transcript(video_id)
```

```python
result
```

```python
transcript = ""
for x in result:
    transcript = transcript + x["text"] + ". "
```

```python
transcript_list = []
for x in result:
    transcript_list.append(x["text"])
```

```python
import json
with open("transcript.json", "w") as write_file:
    json.dump(transcript_list, write_file)
```
